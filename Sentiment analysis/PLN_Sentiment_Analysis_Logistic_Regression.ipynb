```python
# ===============================================
# Exemplo completo de PLN com Classificação de Sentimentos
# ===============================================

import re
import nltk
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from gensim.models import Word2Vec
from nltk.corpus import stopwords
import string
```

```python
# ===============================================
# 2. Dataset realista de avaliações
# ===============================================
# 1 = positivo, 0 = negativo
dados = {
    "avaliacao": [

        # POSITIVAS
        "O celular é ótimo, tem uma câmera excelente!",
        "Entrega rápida e produto de boa qualidade.",
        "Muito bom, atendeu todas as minhas expectativas.",
        "Superou minhas expectativas, recomendo demais!",
        "Produto maravilhoso, desempenho excelente.",
        "A bateria dura bastante.",
        "Excelente custo-benefício.",
        "Gostei muito da qualidade.",
        "Produto muito bonito.",
        "Acabamento impecável.",
        "Chegou antes do prazo.",
        "Muito satisfeito com a compra.",
        "Funciona perfeitamente.",
        "Valeu cada centavo.",
        "Fácil de usar.",
        "Tela muito bonita.",
        "Áudio excelente.",
        "Ótima construção.",
        "Produto resistente.",
        "Recomendo sem dúvidas.",
        "Atendimento excelente.",
        "Muito confortável.",
        "Imagem incrível.",
        "Desempenho acima do esperado.",
        "Produto original.",
        "Muito silencioso.",
        "Bem embalado.",
        "Excelente qualidade.",
        "Design moderno.",
        "Carregamento rápido.",
        "Muito eficiente.",
        "Gostei bastante.",
        "Melhor compra que fiz.",
        "Vale muito a pena.",
        "Ótimo desempenho.",
        "Material de excelente qualidade.",
        "Produto confiável.",
        "A instalação foi simples.",
        "Acabou superando minhas expectativas.",
        "Excelente acabamento.",
        "Muito bonito e funcional.",
        "Funciona muito bem.",
        "Produto fantástico.",
        "Experiência excelente.",
        "Recebi exatamente o que esperava.",
        "Muito feliz com a compra.",
        "Qualidade surpreendente.",
        "Recomendo para todos.",
        "Produto nota dez.",
        "Compra excelente.",

        # NEGATIVAS
        "Horrível! Que produto terrível, não recomendo a ninguém.",
        "A bateria dura pouco e esquenta muito. Péssimo!",
        "Não gostei, o material parece frágil e a tela risca fácil.",
        "O som é baixo e a bateria acabou em poucas horas.",
        "Defeituoso, não funcionou desde o primeiro dia.",
        "Muito ruim.",
        "Produto de péssima qualidade.",
        "Dinheiro jogado fora.",
        "Não vale o preço.",
        "Chegou quebrado.",
        "Parou de funcionar rapidamente.",
        "Extremamente decepcionado.",
        "Não recomendo.",
        "Produto frágil.",
        "Tela apresentou defeito.",
        "A bateria descarrega rápido.",
        "Esquenta demais.",
        "Áudio muito ruim.",
        "Imagem horrível.",
        "Acabamento péssimo.",
        "Produto falsificado.",
        "Veio incompleto.",
        "Muito decepcionante.",
        "Travou logo no primeiro uso.",
        "Não funcionou corretamente.",
        "Qualidade inferior.",
        "Experiência ruim.",
        "Material muito fraco.",
        "Chegou danificado.",
        "Assistência técnica ruim.",
        "Muito lento.",
        "Sistema trava constantemente.",
        "Tela apagou sozinha.",
        "Produto veio com defeito.",
        "Não compraria novamente.",
        "Perdi dinheiro.",
        "Atendimento péssimo.",
        "A entrega atrasou muito.",
        "Embalagem danificada.",
        "Funcionamento instável.",
        "A câmera é ruim.",
        "Muito barulhento.",
        "Não atendeu às expectativas.",
        "Qualidade muito baixa.",
        "Arrependi da compra.",
        "Produto inútil.",
        "Pior compra que já fiz.",
        "Não vale a pena.",
        "Completamente insatisfeito.",
        "Produto horrível."

    ],

    "sentimento": [1]*50 + [0]*50
}

df = pd.DataFrame(dados)
print(df)
```

```python
# ===============================================
# 3. Função de pré-processamento
# ===============================================
def preprocessar_texto(texto):
    texto = texto.lower()
    texto = texto.replace('\n', ' ')
    texto = re.sub(r'[^a-zA-Zà-úÀ-Ú\s]', '', texto)
    tokens = nltk.word_tokenize(texto, language='portuguese')
    tokens = [t for t in tokens if t not in string.punctuation]
    stop_words = set(stopwords.words('portuguese'))
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    return " ".join(tokens)

df["texto_limpo"] = df["avaliacao"].apply(preprocessar_texto)
print(df["texto_limpo"])
```

```python
# ===============================================
# 4. Representações vetoriais (Bag-of-Words e TF-IDF)
# ===============================================
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["texto_limpo"])
y = df["sentimento"]
print(y)
print(X)
```

```python
# ===============================================
# 5. Divisão de dados (treino e teste)
# ===============================================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
# ===============================================
# 6. Aplicação do Machine Learning: modelo de classificação (Regressão Logística)
# ===============================================
modelo = LogisticRegression(max_iter=1000)
modelo.fit(X_train, y_train)
print(modelo)
```

```python
# ===============================================
# 7. Avaliação do modelo
# ===============================================
y_pred = modelo.predict(X_test)
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred, target_names=["Negativo", "Positivo"]))

print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
```

```python
# ===============================================
# 8. Teste com novas avaliações
# ===============================================
novas_avaliacoes = [
    "A câmera é incrível e a bateria dura o dia todo.",
    "O produto veio quebrado e não liga mais.",
    "Entrega rápida, mas o desempenho é ruim.",
    "Excelente custo-benefício, estou muito satisfeito!"
]

def sentimentos(novas_avaliacoes):
    # Pré-processar e transformar
    novas_proc = [preprocessar_texto(a) for a in novas_avaliacoes]
    X_novas = vectorizer.transform(novas_proc)
    predicoes = modelo.predict(X_novas)

    # Exibir resultados
    print("\nResultados em novas avaliações:")
    for texto, pred in zip(novas_avaliacoes, predicoes):
        sentimento = "Positivo 😀" if pred == 1 else "Negativo 😞"
        print(f"- {texto}: {sentimento}")

sentimentos(novas_avaliacoes)
```

```python
# ===============================================
# 9. Word2Vec (para análise semântica complementar)
# ===============================================
tokens = [t.split() for t in df["texto_limpo"]]
modelo_w2v = Word2Vec(sentences=tokens, vector_size=50, window=3, min_count=1, sg=1)

print("\nPalavras mais similares a 'ótimo':")
try:
    print(modelo_w2v.wv.most_similar('ótimo'))
except KeyError:
    print("A palavra 'ótimo' não está no vocabulário.")
```

```python
# ===============================================
# 10. Teste personalizado
# ===============================================
while True:
    frase = input("Digite uma frase (ENTER para sair): ")
    if frase == "":
        break
    sentimentos([frase])
```
