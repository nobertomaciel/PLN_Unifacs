```python
import pandas as pd
from transformers import pipeline
```

```python
sentimentos = pipeline(
    "text-classification",
    model="lxyuan/distilbert-base-multilingual-cased-sentiments-student",
    truncation=True
)
```

```python
frase = "Gostei muito do produto."
resultado = sentimentos(frase)
print(resultado)
```

```python
frases = [
    "Gostei muito do produto.",
    "O atendimento foi excelente.",
    "A câmera é fantástica.",
    "Não gostei do serviço.",
    "Muito ruim.",
    "O celular estragou em dois dias.",
    "Excelente custo-benefício.",
    "Jamais compraria novamente."
]

for frase in frases:
    r = sentimentos(frase)[0]
    print(f"{frase}")
    print(f"Sentimento: {r['label']}")
    print(f"Confiança: {r['score']:.3f}")
    print()
```

```python
avaliacoes = [
  "Produto maravilhoso.",
  "Péssima qualidade.",
  "Gostei bastante.",
  "Não recomendo.",
  "Entrega muito rápida.",
  "A bateria dura pouco."
]

for avaliacao in avaliacoes:
    resultado = sentimentos(avaliacao)[0]
    print("-"*60)
    print(avaliacao)
    print(resultado)
```

```python
texto = "O notebook é muito bom, mas a bateria deixa a desejar."
r = sentimentos(texto)[0]
print("Texto:")
print(texto)
print()
print("Classe:", r["label"])
print("Confiança:", round(r["score"]*100,2),"%")
```

```python
while True:
    frase = input("Digite uma frase (ENTER para sair): ")
    if frase == "":
        break
    r = sentimentos(frase)[0]
    print()
    print("Sentimento:", r["label"])
    print("Confiança:", round(r["score"]*100,2),"%")
    print()
```
